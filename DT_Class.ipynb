```python
import numpy as np 
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
x=pd.DataFrame(iris['data'],columns=['Sepal Length in cm','Sepal Width in cm ','Petal Length in cm ','Petal Width in cm '])
```

```python
x
```

```python
y=iris['target']
```

```python
y
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)   
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report   
print("Accuracy:",accuracy_score(y_test,y_pred))
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
print("Classification Report:\n",classification_report(y_test,y_pred))
plt.figure(figsize=(20,10))
plot_tree(
    model,
    feature_names=x.columns,
    class_names=iris.target_names,
    filled=True,
    rounded=True
)
plt.show()
```

```python
# Decision Tree Classifier with Prepuning and Hyperparameter Tuning.
```

```python
param = {
    'criterion': ['gini','entropy','log_loss'],
    'splitter': ['best','random'],
    'max_depth': [1,2,3,4,5],
    'max_features': ['auto','sqrt','log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model=DecisionTreeClassifier()
```

```python
grid=GridSearchCV(model,param_grid=param,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
y_pred
```

```python
print('confusion_matrix:\n', confusion_matrix(y_test,y_pred))
print('classification_report:\n', classification_report(y_test,y_pred))
```
